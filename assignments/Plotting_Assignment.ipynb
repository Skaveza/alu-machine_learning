```python
plt.figure(figsize=(10, 6))
plt.hist(df['co2'], bins=5, edgecolor='black')
plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['make'], bins=15, edgecolor='black')
plt.title('Distribution of make')
plt.xlabel('make')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['pv2'], bins=10, edgecolor='black')
plt.title('Distribution of pv2')
plt.xlabel('pv2')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Select all columns (including encoded and original numeric columns)
columns = ['id', 'year', 'cylinders', 'displ', 'pv2', 'pv4', 'city', 'UCity', 'highway', 'UHighway', 'comb', 'co2', 'feScore', 'ghgScore']


corr_matrix = df[columns].corr()

plt.figure(figsize = (10,8))
sns.heatmap(corr_matrix, cmap = 'RdBu', vmin = -1, vmax = 1, center = 0)
plt.show()
```

```python
#The intesity of the colors shows the strength of the correlations
#There is a strong positive correlation between each variable and itself
# There is a strong negative correlation between co2 and ghgscore, feScore,comb,UHighway,highway,UCity and city, indicating that as they increase, the level of co2 decrease.
# If the target was the ghgScore I would remove the id and year columns because they have a minimal effect.
```
